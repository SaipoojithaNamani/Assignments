```python
file_name='/content/pima-indians-diabetes.data.csv'
names=['preg','plas','pres','skin','test','mass','pedi','age','class']
dataframe=read_csv(file_name,names=names)
dataframe.head()
```

```python
lis=[1,2,3,5,6,7,8,12,13]
print(lis[0:8])
```

```python
array=dataframe.values
X=array[:,0:8] # Select all rows and columns from 0 to 7
Y=array[:,8]
```

```python
num_trees=100
seed=7
k_fold=KFold(n_splits=10,random_state=seed,shuffle=True)
model=RandomForestClassifier(n_estimators=num_trees,random_state=seed)
k_fold
```

```python
result=cross_val_score(model,X,Y,cv=k_fold)
print(result.mean())
```
